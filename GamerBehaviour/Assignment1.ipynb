```python
import matplotlib.pyplot as plt
import numpy as np  # Import numpy for trigonometric functions

# Assuming 'df2' is your DataFrame and it contains a 'frequency' column
# Group by 'frequency' column (assumed to contain the number of times each player appeared)
player_frequency = df2['frequency'].value_counts()

# Extract labels and sizes for the pie chart
labels = [f'{int(freq)} Times' for freq in player_frequency.index]
sizes = player_frequency.values

# Create a pie chart with custom styling
fig, ax = plt.subplots()

# Customize the pie chart
colors = plt.cm.Paired(range(len(sizes)))  # Use a colormap for different colors
explode = [0.1 if i == 0 else 0 for i in range(len(sizes))]  # Highlight the first slice

# Plotting the pie chart without percentage values inside
wedges, texts = ax.pie(sizes, explode=explode, colors=colors,
                       startangle=90, shadow=True, labeldistance=1.2)

# Add percentage values outside the pie chart
for i, wedge in enumerate(wedges):
    # Calculate the angle for the current wedge to place the percentage
    angle = (wedge.theta2 - wedge.theta1) / 2. + wedge.theta1
    x = 1.2 * wedge.r * np.cos(np.radians(angle))  # Use numpy for cosine
    y = 1.2 * wedge.r * np.sin(np.radians(angle))  # Use numpy for sine
    # Display the percentage value
    plt.text(x, y, f'{sizes[i] / sum(sizes) * 100:.1f}%', ha='left', va='center', fontsize=8)

# Draw a circle at the center to make it a donut chart
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig.gca().add_artist(centre_circle)

# Ensure the pie chart is drawn as a circle
ax.axis('equal')

# Add a legend instead of labels on the pie chart
ax.legend(wedges, labels, title="Number of Appearances", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))

# Add title
plt.title("Proportion of Players by Number of Entries", fontsize=16, weight='bold', pad=30)


# Display the chart
plt.show()
```

```python
from sklearn.multiclass import OneVsRestClassifier
from sklearn.preprocessing import label_binarize
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt
import numpy as np
from scipy.special import logsumexp

# Binarize y for all classes
y_test_bin = label_binarize(y_test, classes=[0, 1, 2])  # Adjust 'classes' to match your specific class labels
n_classes = y_test_bin.shape[1]

# Function to plot multiclass ROC with error handling
def plot_multiclass_roc(models, X_test, y_test_bin):
    plt.figure(figsize=(10, 8))
    for model_name, model in models.items():
        try:
            # Use OneVsRest strategy for multiclass
            ovr = OneVsRestClassifier(model)
            ovr.fit(X_train, y_train)
            
            # Check if the underlying model supports decision_function or predict_proba
            if hasattr(ovr.estimator, "decision_function"):
                y_proba = ovr.decision_function(X_test)
            elif hasattr(ovr.estimator, "predict_proba"):
                y_proba = ovr.predict_proba(X_test)
            else:
                raise AttributeError(f"The model {model_name} does not support decision_function or predict_proba.")
            
            # Calculate ROC for each class
            for i in range(n_classes):
                fpr, tpr, _ = roc_curve(y_test_bin[:, i], y_proba[:, i])
                roc_auc = auc(fpr, tpr)
                plt.plot(fpr, tpr, lw=2, label=f'{model_name} (Class {i}) (AUC = {roc_auc:.2f})')
        
        except FloatingPointError:
            print(f"Skipping model {model_name} due to FloatingPointError.")
        except Exception as e:
            print(f"Error with model {model_name}: {e}")
    
    plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curves for Multiclass Classification')
    plt.legend(loc='lower right')
    plt.show()

# Plot ROC curve for multiclass
plot_multiclass_roc(models, X_test, y_test_bin)
```
